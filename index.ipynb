```python
# Step 1: Install necessary libraries
!pip install transformers sentencepiece --quiet

# Step 2: Import and load the model
from transformers import T5ForConditionalGeneration, T5Tokenizer

model_name = "google/flan-t5-small"
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)

# Step 3: Function to generate a product description
def generate_product_description(product_name):
    input_text = f"Write a compelling product description for: {product_name}"
    input_ids = tokenizer(input_text, return_tensors="pt").input_ids
    outputs = model.generate(input_ids, max_length=100, num_beams=5, early_stopping=True)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# Step 4: Take one-time user input and generate a description
user_product = input("Enter a product name to get a description: ")
description = generate_product_description(user_product)
print("\n🛍️ Generated Product Description:", description)
```
